```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
fuel_counts = data['Fuel_Type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count fuel types
fuel_counts = data['Fuel_Type'].value_counts()

# Combine rare types into "Other"
threshold = 0.02  # group anything <2% into "Other"
total = fuel_counts.sum()
fuel_counts_combined = fuel_counts[fuel_counts / total > threshold]
other_count = fuel_counts[fuel_counts / total <= threshold].sum()

if other_count > 0:
    fuel_counts_combined['Other'] = other_count

# Plot
plt.figure(figsize=(7, 7))
plt.pie(fuel_counts_combined,
        labels=fuel_counts_combined.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Set2.colors)
plt.title('Fuel Type Distribution (Grouped)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Transmission', data=data, palette='Set2')
plt.title('Transmission Type Count')
plt.ylabel('Count')
plt.xlabel('Transmission Type')
plt.show()
```
